```python
def loadTXTData(filename):
	txtlines = []
	with open(file=filename, mode='r', encoding='utf-8') as txtData:
		lines = txtData.readlines()
		for line in lines:
			lineData = line.strip()
			if lineData == "\n":
				continue
			else:
				txtlines.append(lineData)
	return txtlines
```

```python
dictPath = 'data/dict.txt'
```

```python
print(loadTXTData(dictPath)[0])
```

```python
dictdata = loadTXTData(dictPath)[0]
```

```python
words = []
for d in dictdata:
    words.append(d)

words
```

```python
import pandas as pd

df = pd.DataFrame(words)

df.to_excel('words.xlsx')
```
